```python
# Persamaan garis yang melewati 2 titik
# Garis pertama melewati (a,b) dan (c,d)
# Garis kedua melewati (e,f) dan (g,h)
# Terdapat 3 kemungkinan dari kedua persamaan garis ini:
# 1. Terdapat perpotongan di satu titik (gradien kedua garis tidak sama)
# 2. Terdapat tak berhingga banyaknya solusi (kedua garis berhimpitan) (kedua garis sejajar)
# 3. Tidak terdapat perpotongan garis (tidak terdapat solusi) (kedua garis sejajar)

import numpy as np
import matplotlib.pyplot as plt

a = float(input("Nilai x dari titik pertama dari garis pertama adalah "))
b = float(input("Nilai y dari titik pertama dari garis pertama adalah "))
c = float(input("Nilai x dari titik kedua dari garis pertama adalah "))
d = float(input("Nilai y dari titik kedua dari garis pertama adalah "))

e = float(input("Nilai x dari titik pertama dari garis kedua adalah "))
f = float(input("Nilai y dari titik pertama dari garis kedua adalah "))
g = float(input("Nilai x dari titik kedua dari garis kedua adalah "))
h = float(input("Nilai y dari titik kedua dari garis kedua adalah "))

m_1 = (d-b)/(c-a) # Gradien garis pertama
m_2 = (h-f)/(g-e) # Gradien garis kedua

# Persamaan garis pertama adalah y=m_1(x-a)+b=m_1x-m_1a+b
# Persamaan garis kedua adalah y=m_2(x-e)+f=m_2x-m_2e+f

c_1 = -m_1*a+b # Titik potong garis pertama terhadap sumbu y
c_2 = -m_2*e+f # titik potong garis kedua terhadap sumbu y garis kedua

print("Persamaan garis pertama adalah y={}x+{}".format(m_1,c_1))
print("Persamaan garis kedua adalah y={}x+{}".format(m_2,c_2))

if m_1==m_2:
    #print("Kedua garis sejajar sehingga kemungkinannya adalah tidak terdapat solusi atau terdapat tak berhingga banyaknya solusi")
    if c_1 == c_2:
        print("Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi")
        x=np.linspace(-5,5,100)
        plt.plot(x,m_1*(x-a)+b,color='red')
        plt.plot(x,m_2*(x-e)+f,color='blue')
        plt.show()
    else:
        print("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
        x=np.linspace(-5,5,100)
        plt.plot(x,m_1*(x-a)+b,color='red')
        plt.plot(x,m_2*(x-e)+f,color='blue')
        plt.show()
else:
    print("Akan terdapat satu titik potong")
    tp_x = (-m_2*e+f+m_1*a-b)/(m_1-m_2)
    tp_y = m_1*(tp_x-a)+b
    print("Titik potong kedua garis adalah ({},{})".format(tp_x,tp_y))
    
    x=np.linspace(tp_x-5,tp_x+5,100)
    plt.plot(x,m_1*(x-a)+b,color='red')
    plt.plot(x,m_2*(x-e)+f,color='blue')
    plt.axvline(x=0,color='grey')
    plt.axhline(y=0,color='grey')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Validasi input untuk memastikan nilai numerik yang dimasukkan
def validate_input(prompt):
    while True:
        try:
            value = float(input(prompt))
            return value
        except ValueError:
            print("Masukkan harus berupa angka!")

# Mendapatkan input untuk titik-titik dari kedua garis
def get_points():
    print("Masukkan koordinat untuk titik-titik:")
    a = validate_input("Nilai x dari titik pertama dari garis pertama: ")
    b = validate_input("Nilai y dari titik pertama dari garis pertama: ")
    c = validate_input("Nilai x dari titik kedua dari garis pertama: ")
    d = validate_input("Nilai y dari titik kedua dari garis pertama: ")
    e = validate_input("Nilai x dari titik pertama dari garis kedua: ")
    f = validate_input("Nilai y dari titik pertama dari garis kedua: ")
    g = validate_input("Nilai x dari titik kedua dari garis kedua: ")
    h = validate_input("Nilai y dari titik kedua dari garis kedua: ")
    return a, b, c, d, e, f, g, h

# Menghitung gradien dari garis yang melewati dua titik
def calculate_gradient(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1)

# Menghitung nilai titik potong dari dua garis
def calculate_intersection(m1, m2, c1, c2):
    x = (c2 - c1) / (m1 - m2)
    y = m1 * x + c1
    return x, y

# Menampilkan plot dari dua garis
def plot_lines(a, b, c, d, e, f, g, h, m1, m2, tp_x=None, tp_y=None):
    x_values = np.linspace(min(a, c, e, g) - 5, max(a, c, e, g) + 5, 100)
    plt.plot(x_values, m1 * (x_values - a) + b, label="Garis Pertama")
    plt.plot(x_values, m2 * (x_values - e) + f, label="Garis Kedua")
    if tp_x is not None and tp_y is not None:
        plt.scatter(tp_x, tp_y, color='red', label="Titik Potong")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Plot Garis')
    plt.axvline(0, color='black', lw=0.5)
    plt.axhline(0, color='black', lw=0.5)
    plt.legend()
    plt.grid(True)
    plt.show()

def main():
    print("Program Untuk Mencari Titik Potong Dua Garis")
    a, b, c, d, e, f, g, h = get_points()

    # Menghitung gradien masing-masing garis
    m1 = calculate_gradient(a, b, c, d)
    m2 = calculate_gradient(e, f, g, h)

    # Menghitung konstanta c pada persamaan garis y = mx + c
    c1 = b - m1 * a
    c2 = f - m2 * e

    print("Persamaan garis pertama adalah y = {:.3f}x + {:.3f}".format(m1, c1))
    print("Persamaan garis kedua adalah y = {:.3f}x + {:.3f}".format(m2, c2))

    # Menentukan apakah garis-garis tersebut berpotongan
    if m1 == m2:
        if c1 == c2:
            print("Kedua garis berhimpitan, terdapat tak berhingga banyaknya solusi.")
        else:
            print("Kedua garis sejajar, tidak terdapat titik potong antara kedua garis.")
    else:
        # Menghitung titik potong jika garis-garis tersebut berpotongan
        tp_x, tp_y = calculate_intersection(m1, m2, c1, c2)
        print("Titik potong kedua garis adalah ({:.3f}, {:.3f})".format(tp_x, tp_y))
        plot_lines(a, b, c, d, e, f, g, h, m1, m2, tp_x, tp_y)

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def input_point(prompt):
    x = float(input(f"Nilai x {prompt}: "))
    y = float(input(f"Nilai y {prompt}: "))
    return x, y

def calculate_slope(x1, y1, x2, y2):
    if x1 == x2:
        return None  # Slope is undefined (vertical line)
    return (y2 - y1) / (x2 - x1)

def calculate_intercept(x, y, slope):
    if slope is None:
        return None  # Vertical line has no y-intercept
    return y - slope * x

# Input points
a, b = input_point("dari titik pertama dari garis pertama")
c, d = input_point("dari titik kedua dari garis pertama")

e, f = input_point("dari titik pertama dari garis kedua")
g, h = input_point("dari titik kedua dari garis kedua")

# Calculate slopes and intercepts
m_1 = calculate_slope(a, b, c, d)
m_2 = calculate_slope(e, f, g, h)

c_1 = calculate_intercept(a, b, m_1)
c_2 = calculate_intercept(e, f, m_2)

# Display equations
if m_1 is not None:
    print(f"Persamaan garis pertama adalah y = {m_1}x + {c_1}")
else:
    print(f"Persamaan garis pertama adalah x = {a}")

if m_2 is not None:
    print(f"Persamaan garis kedua adalah y = {m_2}x + {c_2}")
else:
    print(f"Persamaan garis kedua adalah x = {e}")

# Determine relationship between lines
if m_1 == m_2:
    if c_1 == c_2:
        print("Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi.")
        x = np.linspace(-10, 10, 100)
        if m_1 is not None:
            plt.plot(x, m_1 * x + c_1, color='red', label='Garis 1 dan Garis 2 (Berhimpitan)')
        else:
            plt.axvline(x=a, color='red', label='Garis 1 dan Garis 2 (Berhimpitan)')
    else:
        print("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
        x = np.linspace(-10, 10, 100)
        if m_1 is not None:
            plt.plot(x, m_1 * x + c_1, color='red', label='Garis 1')
            plt.plot(x, m_2 * x + c_2, color='blue', label='Garis 2')
        else:
            plt.axvline(x=a, color='red', label='Garis 1')
            plt.axvline(x=e, color='blue', label='Garis 2')
else:
    if m_1 is not None and m_2 is not None:
        tp_x = (c_2 - c_1) / (m_1 - m_2)
        tp_y = m_1 * tp_x + c_1
        print(f"Titik potong kedua garis adalah ({tp_x}, {tp_y})")
        x = np.linspace(tp_x - 5, tp_x + 5, 100)
        plt.plot(x, m_1 * x + c_1, color='red', label='Garis 1')
        plt.plot(x, m_2 * x + c_2, color='blue', label='Garis 2')
        plt.plot(tp_x, tp_y, 'go', label='Titik Potong')
    else:
        if m_1 is None:
            tp_x = a
            tp_y = m_2 * tp_x + c_2
        else:
            tp_x = e
            tp_y = m_1 * tp_x + c_1
        print(f"Titik potong kedua garis adalah ({tp_x}, {tp_y})")
        x = np.linspace(-10, 10, 100)
        if m_1 is not None:
            plt.plot(x, m_1 * x + c_1, color='red', label='Garis 1')
        else:
            plt.axvline(x=a, color='red', label='Garis 1')
        if m_2 is not None:
            plt.plot(x, m_2 * x + c_2, color='blue', label='Garis 2')
        else:
            plt.axvline(x=e, color='blue', label='Garis 2')
        plt.plot(tp_x, tp_y, 'go', label='Titik Potong')

# Plotting the axes
plt.axvline(x=0, color='grey', linestyle='--', linewidth=0.5)
plt.axhline(y=0, color='grey', linestyle='--', linewidth=0.5)
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Visualisasi Garis Sejajar')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def persamaan_garis(a, b, c, d):
    m = (d - b) / (c - a)
    c_ = -m * a + b
    return m, c_

def plot_garis(a, b, m, c, color):
    x = np.linspace(a - 5, c + 5, 100)
    plt.plot(x, m * (x - a) + b, color=color)

def titik_potong(m1, c1, m2, c2):
    if m1 == m2:
        if c1 == c2:
            print("Kedua garis berhimpitan, terdapat tak berhingga banyaknya solusi.")
        else:
            print("Kedua garis sejajar, tidak terdapat titik potong antara kedua garis.")
    else:
        tp_x = (c2 - c1) / (m1 - m2)
        tp_y = m1 * tp_x + c1
        print("Titik potong kedua garis adalah ({:.2f}, {:.2f})".format(tp_x, tp_y))
        return tp_x, tp_y

def main():
    print("Program Untuk Mencari Titik Potong Dua Garis")
    print("Masukkan koordinat untuk dua titik pada masing-masing garis.")
    
    a = float(input("Nilai x dari titik pertama dari garis pertama adalah: "))
    b = float(input("Nilai y dari titik pertama dari garis pertama adalah: "))
    c = float(input("Nilai x dari titik kedua dari garis pertama adalah: "))
    d = float(input("Nilai y dari titik kedua dari garis pertama adalah: "))

    e = float(input("Nilai x dari titik pertama dari garis kedua adalah: "))
    f = float(input("Nilai y dari titik pertama dari garis kedua adalah: "))
    g = float(input("Nilai x dari titik kedua dari garis kedua adalah: "))
    h = float(input("Nilai y dari titik kedua dari garis kedua adalah: "))

    m1, c1 = persamaan_garis(a, b, c, d)
    m2, c2 = persamaan_garis(e, f, g, h)

    print("Persamaan garis pertama adalah y = {:.3f}x + {:.3f}".format(m1, c1))
    print("Persamaan garis kedua adalah y = {:.3f}x + {:.3f}".format(m2, c2))

    plot_garis(a, b, m1, c1, 'red')
    plot_garis(e, f, m2, c2, 'blue')

    tp = titik_potong(m1, c1, m2, c2)
    if tp:
        plt.scatter(tp[0], tp[1], color='black', label="Titik Potong")
        plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Plot Garis')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```

```python
# 1. Garis melewati 2 titik
# 2. Garis diketahui gradien dan titik yang dilewati
# 3. Megikuti persamaan garis ax+by=c dan dx+ey=f

# Akan dicari titik potong dari dua garis yg memenuhi salah satu dari ketiga kasus di atas

import numpy as np
import matplotlib.pyplot as plt

def garis_melewati_dua_titik():
    print("Masing-masing garis akan melewati dua titik yang diketahui.")
    a = float(input("Nilai x dari titik pertama dari garis pertama adalah: "))
    b = float(input("Nilai y dari titik pertama dari garis pertama adalah: "))
    c = float(input("Nilai x dari titik kedua dari garis pertama adalah: "))
    d = float(input("Nilai y dari titik kedua dari garis pertama adalah: "))

    e = float(input("Nilai x dari titik pertama dari garis kedua adalah: "))
    f = float(input("Nilai y dari titik pertama dari garis kedua adalah: "))
    g = float(input("Nilai x dari titik kedua dari garis kedua adalah: "))
    h = float(input("Nilai y dari titik kedua dari garis kedua adalah: "))

    m_1 = (d - b) / (c - a)  # Gradien garis pertama
    m_2 = (h - f) / (g - e)  # Gradien garis kedua

    c_1 = -m_1 * a + b  # Titik potong garis pertama terhadap sumbu y
    c_2 = -m_2 * e + f  # Titik potong garis kedua terhadap sumbu y garis kedua

    print("Persamaan garis pertama adalah y = {:.3f}x + {:.3f}".format(m_1, c_1))
    print("Persamaan garis kedua adalah y = {:.3f}x + {:.3f}".format(m_2, c_2))

    if m_1 == m_2:
        if c_1 == c_2:
            print("Kedua garis berhimpitan, terdapat tak berhingga banyaknya solusi.")
            plot_garis(a, b, c, d, e, f, g, h, m_1, m_2)
        else:
            print("Kedua garis tidak mungkin berpotongan, tidak ada solusi.")
            plot_garis(a, b, c, d, e, f, g, h, m_1, m_2)
    else:
        tp_x = (-m_2 * e + f - m_1 * a + b) / (m_1 - m_2)
        tp_y = m_1 * (tp_x - a) + b
        print("Titik potong kedua garis adalah ({:.3f}, {:.3f})".format(tp_x, tp_y))
        plot_garis(a, b, c, d, e, f, g, h, m_1, m_2, tp_x, tp_y)

def garis_diketahui_gradien_dan_titik():
    print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
    m_1 = float(input("Gradien dari garis pertama adalah: "))
    a = float(input("Nilai x dari titik yang dilewati oleh garis pertama adalah: "))
    b = float(input("Nilai y dari titik yang dilewati oleh garis pertama adalah: "))
    print("Persamaan garis pertama adalah y = {:.3f}x + {:.3f}".format(m_1, b))

    m_2 = float(input("Gradien dari garis kedua adalah: "))
    c = float(input("Nilai x dari titik yang dilewati oleh garis kedua adalah: "))
    d = float(input("Nilai y dari titik yang dilewati oleh garis kedua adalah: "))
    print("Persamaan garis kedua adalah y = {:.3f}x + {:.3f}".format(m_2, d))

    c_1 = -m_1 * a + b
    c_2 = -m_2 * c + d

    if m_1 == m_2:
        if c_1 == c_2:
            print("Kedua garis sejajar, terdapat tak berhingga banyaknya solusi.")
        else:
            print("Tidak terdapat titik potong antara kedua garis.")
    else:
        print("Kedua garis akan berpotongan di satu titik.")

def garis_mengikuti_persamaan():
    print("Kedua garis memenuhi persamaan berbentuk ax + by = c dan dx + ey = f.")
    a = float(input("Koefisien dari x pada garis pertama adalah: "))
    b = float(input("Koefisien dari y pada garis pertama adalah: "))
    c = float(input("Konstanta dari garis pertama adalah: "))

    d = float(input("Koefisien dari x pada garis kedua adalah: "))
    e = float(input("Koefisien dari y pada garis kedua adalah: "))
    f = float(input("Konstanta dari garis kedua adalah: "))

    m_1 = -a / b
    m_2 = -d / e

    c_1 = c / b
    c_2 = f / e

    if m_1 == m_2:
        if c_1 == c_2:
            print("Kedua garis berhimpitan.")
        else:
            print("Kedua garis sejajar.")
    else:
        print("Kedua garis akan berpotongan di satu titik.")

def plot_garis(a, b, c, d, e, f, g, h, m_1, m_2, tp_x=None, tp_y=None):
    x = np.linspace(min(a, c, e, g) - 5, max(a, c, e, g) + 5, 100)
    plt.plot(x, m_1 * (x - a) + b, color='red', label="Garis Pertama")
    plt.plot(x, m_2 * (x - e) + f, color='blue', label="Garis Kedua")
    plt.axvline(x=0, color='grey')
    plt.axhline(y=0, color='grey')
    if tp_x is not None and tp_y is not None:
        plt.scatter(tp_x, tp_y, color='black', label="Titik Potong")
    plt.legend(loc='best')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Plot Garis')
    plt.grid(True)
    plt.show()

def main():
    print("Program Untuk Mencari Titik Potong Dua Garis")
    print("Pilih salah satu kasus:")
    print("1. Garis melewati dua titik")
    print("2. Garis diketahui gradien dan titik yang dilewati")
    print("3. Megikuti persamaan garis ax+by=c dan dx+ey=f")

    kasus = int(input("Masukkan nomor kasus: "))

    if kasus == 1:
        garis_melewati_dua_titik()
    elif kasus == 2:
        garis_diketahui_gradien_dan_titik()
    elif kasus == 3:
        garis_mengikuti_persamaan()
    else:
        print("Masukan tidak valid.")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def garis_dari_dua_titik(x1, y1, x2, y2):
    if x1 == x2:
        return None, None, f"x = {x1}"  # Garis vertikal
    m = (y2 - y1) / (x2 - x1)
    b = y1 - m * x1
    return m, b, f"y = {m}x + {b}"

def garis_dari_gradien_titik(m, x, y):
    b = y - m * x
    return m, b, f"y = {m}x + {b}"

def garis_dari_persamaan(a, b, c):
    if b == 0:
        return None, None, f"x = {c / a}"  # Garis vertikal
    m = -a / b
    intercept = c / b
    return m, intercept, f"y = {m}x + {intercept}"

def main():
    print("Pilih kondisi:")
    print("1. Diketahui 2 titik")
    print("2. Diketahui 1 gradien dan 1 titik yang dilewati")
    print("3. Diketahui persamaan ax + by = c")
    pilihan = int(input("Masukkan pilihan (1/2/3): "))

    if pilihan == 1:
        x1 = float(input("Masukkan nilai x dari titik pertama: "))
        y1 = float(input("Masukkan nilai y dari titik pertama: "))
        x2 = float(input("Masukkan nilai x dari titik kedua: "))
        y2 = float(input("Masukkan nilai y dari titik kedua: "))
        m, b, persamaan = garis_dari_dua_titik(x1, y1, x2, y2)
    elif pilihan == 2:
        m = float(input("Masukkan nilai gradien: "))
        x = float(input("Masukkan nilai x dari titik: "))
        y = float(input("Masukkan nilai y dari titik: "))
        m, b, persamaan = garis_dari_gradien_titik(m, x, y)
    elif pilihan == 3:
        a = float(input("Masukkan nilai a: "))
        b = float(input("Masukkan nilai b: "))
        c = float(input("Masukkan nilai c: "))
        m, b, persamaan = garis_dari_persamaan(a, b, c)
    else:
        print("Pilihan tidak valid.")
        return

    print(f"Persamaan garis: {persamaan}")

    # Visualisasi garis
    x = np.linspace(-10, 10, 100)
    if m is not None:
        y = m * x + b
        plt.plot(x, y, label=persamaan)
    else:
        plt.axvline(x=float(persamaan.split('=')[1]), color='red', label=persamaan)

    # Menggambar sumbu x dan y
    plt.axvline(x=0, color='grey', linestyle='--', linewidth=0.5)
    plt.axhline(y=0, color='grey', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Visualisasi Garis')
    plt.grid()
    plt.show()

if __name__ == "__main__":
    main()
```
