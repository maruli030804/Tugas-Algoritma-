```python
import math
import numpy as np

# Definisi koefisien persamaan kuadrat
a = 1
b = 1
c = -1

# Menghitung akar-akar persamaan kuadrat
x1 = (-b + math.sqrt(b**2 - 4*a*c)) / (2*a)
x2 = (-b - math.sqrt(b**2 - 4*a*c)) / (2*a)

print("Nilai x1 yang memenuhi persamaan adalah:", x1)
print("Nilai x2 yang memenuhi persamaan adalah:", x2)

# Menghitung nilai n untuk x1
n1 = np.log((x1**2024 - x1**2025) / (x1 * (1 - x1))) / np.log(x1 * (1 - x1))

# Menghitung nilai n untuk x2
n2 = np.log((x2**2024 - x2**2025) / (x2 * (1 - x2))) / np.log(x2 * (1 - x2))

print("Nilai n yang memenuhi persamaan untuk x1 adalah:", n1)
print("Nilai n yang memenuhi persamaan untuk x2 adalah:", n2)
```

```python
import math

# Mencari akar-akar dari persamaan kuadrat x^2 + x - 1 = 0
def cari_akar():
    a = 1
    b = 1
    c = -1
    D = b**2 - 4*a*c  
    x1 = (-b + math.sqrt(D)) / (2*a)
    x2 = (-b - math.sqrt(D)) / (2*a)
    return x1, x2

# Menghitung nilai n berdasarkan x
def hitung_n(x):
    numerator = x**2024 - x**2025
    denominator = x * (1 - x)
    if denominator == 0:
        return float('inf')  # jika pembaginya nol, hasilnya tak hingga
    n = math.log(numerator / denominator) / math.log(x * (1 - x))
    return n

# Mencari akar-akar persamaan kuadrat
x1, x2 = cari_akar()

# Menghitung nilai n untuk kedua akar
n1 = hitung_n(x1)
n2 = hitung_n(x2)

print("Nilai x1 yang memenuhi persamaan adalah:", x1)
print("Nilai n yang memenuhi persamaan untuk x1 adalah:", n1)

print("Nilai x2 yang memenuhi persamaan adalah:", x2)
print("Nilai n yang memenuhi persamaan untuk x2 adalah:", n2)

```
