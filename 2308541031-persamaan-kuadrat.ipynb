```python
import matplotlib.pyplot as plt
import numpy as np

def persamaan_kuadrat(a, b, c):
    return a, b, c

def titik_puncak(a, b, c):
    x_puncak = -b / (2 * a)
    y_puncak = a * x_puncak**2 + b * x_puncak + c
    return x_puncak, y_puncak

def posisi_titik_puncak(x_puncak):
    if x_puncak < 0:
        return "di kiri sumbu y"
    elif x_puncak == 0:
        return "pada sumbu y"
    else:
        return "di kanan sumbu y"

def kondisi_parabola_memotong_sumbu_y(c):
    return f"Parabola memotong sumbu y di titik (0, {c})"

def macam_akar_persamaan_kuadrat(a, b, c):
    diskriminan = b**2 - 4*a*c
    if diskriminan > 0:
        return "Persamaan kuadrat memiliki dua akar real yang berbeda"
    elif diskriminan == 0:
        return "Persamaan kuadrat memiliki dua akar real yang sama"
    else:
        return "Persamaan kuadrat memiliki dua akar kompleks"

def grafik_kuadrat(a, b, c):
    x = np.linspace(-10, 10, 400)
    y = a * x**2 + b * x + c

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'{a}x^2 + {b}x + {c}')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    
    # Menentukan titik puncak dan sumbu simetri
    x_puncak, y_puncak = titik_puncak(a, b, c)
    plt.plot(x_puncak, y_puncak, 'ro')  # Titik puncak
    plt.axvline(x_puncak, color='r', linestyle='--', label=f'Sumbu Simetri: x = {x_puncak}')
    
    plt.legend()
    plt.title('Grafik Fungsi Kuadrat')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

# Input koefisien dari pengguna
a = float(input("Masukkan koefisien a: "))
b = float(input("Masukkan koefisien b: "))
c = float(input("Masukkan koefisien c: "))

# Menentukan persamaan kuadrat
a, b, c = persamaan_kuadrat(a, b, c)
print(f"Persamaan kuadrat: {a}x^2 + {b}x + {c} = 0")

# Menampilkan grafik fungsi kuadrat
grafik_kuadrat(a, b, c)

# Menentukan titik puncak dan sumbu simetri
x_puncak, y_puncak = titik_puncak(a, b, c)
print(f"Titik puncak: ({x_puncak}, {y_puncak})")
print(f"Sumbu simetri: x = {x_puncak}")

# Menentukan posisi titik puncak
posisi = posisi_titik_puncak(x_puncak)
print(f"Posisi titik puncak: {posisi}")

# Menentukan kondisi parabola memotong sumbu y
kondisi_sumbu_y = kondisi_parabola_memotong_sumbu_y(c)
print(kondisi_sumbu_y)

# Menentukan macam-macam akar persamaan kuadrat
macam_akar = macam_akar_persamaan_kuadrat(a, b, c)
print(macam_akar)
```

```python
import math

def hitung_akar(a, b, c):
    diskriminan = b**2 - 4*a*c
    
    if a == 0:
        print("Koefisien a tidak boleh nol dalam persamaan kuadrat.")
        return None
    
    if diskriminan > 0:
        akar1 = (-b + math.sqrt(diskriminan)) / (2 * a)
        akar2 = (-b - math.sqrt(diskriminan)) / (2 * a)
        return akar1, akar2, "Persamaan kuadrat memiliki dua akar real yang berbeda."
    elif diskriminan == 0:
        akar = -b / (2 * a)
        return akar, akar, "Persamaan kuadrat memiliki dua akar real yang sama."
    else:
        real_part = -b / (2 * a)
        imag_part = math.sqrt(-diskriminan) / (2 * a)
        akar1 = complex(real_part, imag_part)
        akar2 = complex(real_part, -imag_part)
        return akar1, akar2, "Persamaan kuadrat memiliki dua akar kompleks."

# Input koefisien dari pengguna
a = float(input("Masukkan koefisien a: "))
b = float(input("Masukkan koefisien b: "))
c = float(input("Masukkan koefisien c: "))

# Menyelesaikan persamaan kuadrat
hasil = hitung_akar(a, b, c)

if hasil:
    akar1, akar2, pesan = hasil
    print(f"Akar-akar dari persamaan kuadrat: {akar1}, {akar2}")
    print(pesan)

    
```

```python
def persamaan_kuadrat_dari_akar(x1, x2):
    a = 1.0  # Secara umum kita bisa menetapkan a = 1 untuk kemudahan
    b = -(x1 + x2)
    c = x1 * x2
    return a, b, c

def format_persamaan(a, b, c):
    persamaan = f"{a}x^2 "
    if b < 0:
        persamaan += f"- {-b}x "
    else:
        persamaan += f"+ {b}x "
    
    if c < 0:
        persamaan += f"- {-c} = 0"
    else:
        persamaan += f"+ {c} = 0"
    
    return persamaan

def main():
    try:
        a = float(input("Masukkan koefisien a: "))
        b = float(input("Masukkan koefisien b: "))
        c = float(input("Masukkan koefisien c: "))
        
        persamaan = format_persamaan(a, b, c)
        print(f"Persamaan kuadrat yang terbentuk: {persamaan}")
        
    except ValueError:
        print("Input tidak valid. Pastikan Anda memasukkan angka yang benar.")

if __name__ == "__main__":
    main()

    
```
