```python
import numpy as np
import matplotlib.pyplot as plt

# Persamaan garis pertama yang melewati titik (a,b) dan (c,d)
a = 1 # titik a dan b
b = 1
c = 2 # titik c dan d
d = 2

# Gradien garis pertama
m1 = (d - b) / (c - a)
print("Gradien untuk garis pertama adalah", m1)
print("Persamaan garis pertama adalah y1 = {}x + {}".format(m1, b - m1 * a))
# y1 = m1(x-a) + b

# Persamaan garis kedua yang melewati titik (w,x) dan(y,z)
w = 1 # titik w dan x
x = 1
y = 2 # titik y dan z
z = 2

# Gradien garis kedua 
m2 = (z - x) / (y - w)
print('Gradien garis kedua adalah', m2)
print('Persamaan garis kedua adalah y2 = {}x + {}'.format(m2, x - m2 * w))
# y2 = m2(x-w) + x

if m1 == m2:
    print('Berdasarkan pernyataan, dapat dilihat bahwa gradien sama, sehingga dapat disimpulkan bahwa tidak ada solusi atau terdapat banyak solusi.')
else:
    # Titik potong kedua garis pada (tp_x, tp_y)
    tp_x = (m1 * a - b - m2 * w + x ) / (m1 - m2)
    tp_y = m2 * (tp_x - w) + x
    
    print('Titik potongnya adalah ({},{})'.format(tp_x, tp_y))
    
    x = np.arange(tp_x - 10 , tp_x + 10 , 0.01)
    plt.figure(figsize=(5, 5), dpi = 400)
    
    plt.plot(x, m1 * (x - a) + b, color='blue', label='y = {}x + {}'.format(m1, b - m1 * a))
    plt.plot(x, m2 * (x - w) + x, color='black', label='y = {}x + {}'.format(m2, x - m2 * w))
    plt.plot(tp_x, tp_y , color='yellow', marker='o', markersize=10)
    plt.legend(loc='best')
    
    plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt


def persamaan_garis(x1, y1, x2, y2):
    # Mencari gradien 
    m = (y2 - y1) / (x2 - x1)
    
    # Mencari konstanta 
    c = y1 - m * x1
    
    return m, c

# Menggunakan fungsi untuk mencari persamaan garis yang melalui dua titik (x1, y1) dan (x2, y2)
x1 = float(input("Masukkan nilai x1: "))
y1 = float(input("Masukkan nilai y1: "))
x2 = float(input("Masukkan nilai x2: "))
y2 = float(input("Masukkan nilai y2: "))

m, c = persamaan_garis(x1, y1, x2, y2)

print(f"Persamaan garis adalah: y = {m}x + {c}")

# Plot garis
x_values = np.linspace(min(x1, x2) - 1, max(x1, x2) + 1, 100)
y_values = m * x_values + c

plt.plot(x_values, y_values, label=f'y = {m}x + {c}')
plt.scatter([x1, x2], [y1, y2], color='red', label='Titik')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Persamaan Garis')
plt.legend('best')
plt.grid(True)
plt.show()
```

```python
import math


#luas persegi
def hitung_luas_persegi(sisi):
    return sisi ** 2

# Masukkan panjang sisi persegi
sisi_persegi = float(input("Masukkan panjang sisi persegi (cm): "))

# Hitung luas persegi
luas_persegi = sisi_persegi ** 2

print('luas persegi tersebut adalah: ', luas_persegi, ' cm^2')



#luas 1/4 lingkaran dengan jari jari 10 cm
def hitung_luas_1_4_lingkaran(jari_jari):
    return (1/4) * math.pi * jari_jari_r ** 2

# Masukkan jari-jari lingkaran
jari_jari_r = 10  # cm

# Hitung luas 1/4 lingkaran
luas_1_4_lingkaran = hitung_luas_1_4_lingkaran(jari_jari_r)

print("Luas 1/4 lingkaran dengan jari-jari 10 cm adalah:", luas_1_4_lingkaran, "cm^2")



def hitung_luas_segitiga_siku_siku (alas, tinggi):
    return (1/2) * alas * tinggi

#luas daerah yang di arsir (segitiga siku-siku)
alas_segitiga = sisi_persegi
tinggi_segitiga = jari_jari_r - sisi_persegi

#hitung luas daeerah yang di arsir (segitiga siku siku)
luas_segitiga_siku_siku = 1/2 * alas_segitiga * tinggi_segitiga 


print('luas segitiga siku siku tersebut adalah : ', luas_segitiga_siku_siku, 'cm^2')


#selain itu luas daerah yang di arsir(segitiga siku siku) dapat di proleh dari luas 1/4 lingkaran - luas persegi

luas_segitiga = luas_1_4_lingkaran - luas_persegi

print('luas daerah yng diarsir adalah: ', luas_segitiga, 'cm^2')
```
